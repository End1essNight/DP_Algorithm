```python
import os
import pandas as pd
```

```python
folder = r'C:\Users\FeiyuanYang\OneDrive - Cambridge GaN Devices Ltd\Documents\CGD documents\Projects\R0004\dynamic'
file = 'R0004 TP-SW-118 dynamic data.csv'
df = pd.read_csv(os.path.join(folder, file), skiprows=6)

lst = ['' for _ in range(len(df))]
df.insert(0, 'ICeGaN', lst)
df.insert(1, 'Pattern', lst)
df.insert(2, 'Substrate', lst)
df.insert(3, 'Condition', lst)
df.insert(4, 'HS', lst)

for i in range(len(df)):
    comment = df['comment'].iloc[i]
    #print(comment.split(' '))
    icegan = comment.split(' ')[2]
    pattern = comment.split(' ')[4]
    temp = comment.split(' ')[5][:-1]
    if "Substrate" in comment:
        substrate = comment.split('Substrate ')[1].split(',')[0]
    else:
        substrate = ""
    if "," in comment:
        cond = comment.split(', ')[-2]
    else:
        cond = ""

    if "Substrate" in cond or "RT" in cond:
        cond = ''
    elif "RDSON" in comment:
        cond = 'RDSON'
    hs = comment.split(' ')[-1]

    df.loc[i, 'ICeGaN'] = icegan
    df.loc[i, 'Pattern'] = pattern
    df.loc[i, 'Substrate'] = substrate
    df.loc[i, 'Condition'] = cond
    df.loc[i, 'HS'] = hs
```

```python
output_path = os.path.join(folder, 'R0004 TP-SW-118 dynamic data_add_filters.csv')
df.to_csv(output_path, index=False, header=True)
```
