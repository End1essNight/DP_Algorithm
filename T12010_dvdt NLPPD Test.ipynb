```python
import pandas as pd
import os
from algorithms.tss_process_T12010 import TSS_process
from natsort import natsorted
```

```python
folder = r'Z:\DynamicTestData\T12010\dvdt NLPPD Test'
files = os.listdir(folder)
files = natsorted(files)
files = [f for f in files if f.endswith('.tss')]
output_folder = r'C:\Users\FeiyuanYang\OneDrive - Cambridge GaN Devices Ltd\Documents\CGD documents\Projects\T12010\data analysis'
output_name = 'T12010 dvdt NLPPD Test update'
```

```python
output = [
    ['Source file', 'Mark ID', 'Vd', 'Frequency (Hz)', 'MC_Gate', 'Total pulse number', 'Rgon', 'Vdd',
     'Frame', 'Peak Current Is (A)', 'dV/dt avg (V/ns)', 'dV/dt max (V/ns)', 'Peak Power (W)', 'Eoff (uJ)']
]
```

```python
for f in files:
    #print(f)
    if f != '2601A000LR_400V_5kHz_RgONH_100R_MC_Gate0V_30K_Pulses_100Frames_0.845mA.tss':
        continue
    chip = f.split('_')[0]
    vd = f.split('_')[1][:-1]
    frequency = f.split('_')[2][:-2]
    mcg = f.split('MC_Gate')[1].split('V')[0]
    pulse_n = f.split('_Pulses')[0].split('_')[-1]
    if 'RgON' in f:
        rgon = f.split('R_MC')[0].split('_')[-1]
    else:
        rgon = ''
    if 'Vdd' in f:
        vdd = f.split('Vdd')[1].split('V')[0]
    else:
        vdd = '12'

    tss = TSS_process(folder, f)
    peak_current, peak_power, dvdt_avg, dvdt_max, eoff_all = tss.T12010_extraction()
    for i in range(len(peak_current)):
        line = [f[:-4], chip, vd, frequency, mcg, pulse_n, rgon, vdd,
                i, peak_current[i], dvdt_avg[i], dvdt_max[i], peak_power[i], eoff_all[i]]
        output.append(line)
```

```python
output_df = pd.DataFrame(output)
output_path = os.path.join(output_folder, output_name + '.csv')
output_df.to_csv(output_path, index=False, header=False)
```
