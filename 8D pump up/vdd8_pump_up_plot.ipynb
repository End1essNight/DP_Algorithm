```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
folder = r'C:\Users\FeiyuanYang\OneDrive - Cambridge GaN Devices Ltd\Documents\CGD documents\Projects\P42016'
file = 'P42016_Vdd8_pump_up_lsoff.csv'
path = os.path.join(folder, file)
df = pd.read_csv(path)
```

```python
print(df.columns)
```

```python
grouped = df.groupby(['Board ID', 'Variant', 'Vdc', 'Fsw (kHz)', 'dt (ns)', 'Vg', 'Rgon'])

for name, group in grouped:
    plt.plot(group['Pulse number'], group['Vdd8_settled_lsoff'], label='Vdd8_settled_lsoff')
    title = f"{name[0]}_{name[1]}_Vdc{name[2]}_Fsw{name[3]}_dt{name[4]}_Vg{name[5]}_Rgon{name[6]}"
    plt.title(title)
    plt.xlabel('Pulse number')
    plt.ylabel('Vdd8_settled_lsoff')
    plt.grid()
    save_path = os.path.join(r'C:\Users\FeiyuanYang\OneDrive - Cambridge GaN Devices Ltd\Documents\CGD documents\Projects\P42016\VDD8 pump up figure', name[1], f'{title}.jpg')
    plt.savefig(save_path)
    plt.close() # Close to free up memory
```
