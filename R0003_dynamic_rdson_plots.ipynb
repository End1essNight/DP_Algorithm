```python
import matplotlib.pyplot as plt
import os
from pathlib import Path
import pandas as pd
from pathlib import Path
```

```python
def find_file(filename, search_dir):
    for root, dirs, files in os.walk(search_dir):
        if filename in files:
            return root
    return None

def glob_files(base_path, file_type):
    paths = Path(base_path).rglob(file_type)
    parent_folders = []
    file_names = []

    for p in paths:
        parent_folders.append(p.parent)
        file_names.append(p.name)

    return parent_folders, file_names
```

```python
input_folders = [
    r'C:\Users\FeiyuanYang\OneDrive - Cambridge GaN Devices Ltd\Documents\CGD documents\Projects\R0003\data_analysis\R0003 BDS Dynamic Rdson',
]
folder = r'C:\Users\FeiyuanYang\OneDrive - Cambridge GaN Devices Ltd\Documents\CGD documents\Projects\R0003\data_analysis\R0003 BDS Dynamic Rdson Plots'
```

```python
for input_folder in input_folders:
    csv_folders, csv_files = glob_files(input_folder, '*.csv')
    for i in range(len(csv_files)):
        csv_file_path = os.path.join(csv_folders[i], csv_files[i])
        df = pd.read_csv(csv_file_path, skiprows=7)
        plt.figure(figsize=(10, 5))
        name = csv_files[i][:-4]
        name = name.replace('2248', '2526')
        chip = name.split('_')[1]
        path = Path(os.path.join(folder, f'{chip}_Rdson_plots'))
        path.mkdir(parents=True, exist_ok=True)
        if csv_files[i] == 'MultiDynRdson_2248A0009C_Vhv400_It1_Np3000_test_config_1.csv': 
            plt.plot(df['Frame'][1:], df['RDSON (mOhm)'][1:])
        else:
            plt.plot(df['Frame'], df['RDSON (mOhm)'])
        plt.xlabel('Pulse number')
        plt.ylabel('RDSON (mOhm)')
        plt.xlim([0, 3000])
        plt.title(f'{csv_files[i][:-4]}')
        plt.grid()
        plt.savefig(os.path.join(path, f'{name}.png'))
        plt.close()
```
