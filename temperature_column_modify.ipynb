```python
import os
import pandas as pd
```

```python
folder = r'C:\Users\FeiyuanYang\OneDrive - Cambridge GaN Devices Ltd\Documents\CGD documents\Projects\R0003\data_analysis'
file = 'R0003 OctoGaN DP Socket test.csv'
df = pd.read_csv(os.path.join(folder, file))
```

```python
df.insert(loc=df.columns.get_loc('L (uH)'), column='Temperature (C)', value='')

for idx in range(len(df)):
    comment = df.loc[idx, 'Comment']
    if 'RT' in comment:
        df.loc[idx, 'Temperature (C)'] = 25
    elif 'HT' in comment or '125' in comment:
        df.loc[idx, 'Temperature (C)'] = 125
    else:
        df.loc[idx, 'Temperature (C)'] = 25
```

```python
output_path = os.path.join(folder, file)
df.to_csv(output_path, index=False, header=True)
```
